```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
duration = 2.0  # Total duration in seconds
sample_rate = 44100  # Sample rate in Hz

# Time vector
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

# Modulating frequency parameters
modulating_frequency = 2.0  # Modulating frequency in Hz
carrier_frequency = 440.0  # Carrier frequency in Hz
frequency_deviation = 200.0  # Frequency deviation in Hz

# Create the frequency-modulated sinusoidal signal
signal = np.sin(2 * np.pi * carrier_frequency * t + 
                frequency_deviation * np.sin(2 * np.pi * modulating_frequency * t))

# Generate and plot the spectrogram
plt.specgram(signal, Fs=sample_rate, NFFT=1024, noverlap=512, cmap='viridis')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Frequency-Modulated Sinusoid Spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
duration = 2.0  # Total duration in seconds
sample_rate = 44100  # Sample rate in Hz

# Exponential modulation parameters
decrease_duration = 0.5  # Duration for decreasing to 0 in seconds
increase_duration = 0.5  # Duration for increasing back in seconds

# Calculate the number of samples for the exponential modulation
decrease_samples = int(decrease_duration * sample_rate)
increase_samples = int(increase_duration * sample_rate)
total_samples = decrease_samples + increase_samples

# Create an exponential curve for decreasing and increasing
decrease_curve = np.exp(-5 * np.linspace(0, 1, decrease_samples))
increase_curve = np.exp(-5 * np.linspace(0, 1, increase_samples)[::-1])

# Concatenate and repeat the curves to match the total duration
exponential_curve = np.concatenate([decrease_curve, increase_curve])
exponential_curve = np.tile(exponential_curve, int(np.ceil(total_samples / len(exponential_curve))))

# Create the time vector
t = np.linspace(0, duration, len(exponential_curve), endpoint=False)

# Plot the exponential curve
plt.figure(figsize=(8, 4))
plt.plot(t, exponential_curve, color='b')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Exponential Curve')
plt.gca().set_facecolor('k')  # Set background color to match spectrogram

# Show the same color scheme as the spectrogram
plt.gca().invert_yaxis()
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Parameters
duration = 2.0  # Total duration in seconds
sample_rate = 44100  # Sample rate in Hz

# Exponential modulation parameters
decrease_duration = 0.5  # Duration for decreasing to 0 in seconds
increase_duration = 0.5  # Duration for increasing back in seconds

# Calculate the number of samples for the exponential modulation
decrease_samples = int(decrease_duration * sample_rate)
increase_samples = int(increase_duration * sample_rate)
total_samples = decrease_samples + increase_samples

# Create an exponential curve for decreasing and increasing
decrease_curve = np.exp(-5 * np.linspace(0, 1, decrease_samples))
increase_curve = np.exp(-5 * np.linspace(0, 1, increase_samples)[::-1])

# Concatenate and repeat the curves to match the total duration
exponential_curve = np.concatenate([decrease_curve, increase_curve])
exponential_curve = np.tile(exponential_curve, int(np.ceil(total_samples / len(exponential_curve)))

# Create the time vector
t = np.linspace(0, duration, len(exponential_curve), endpoint=False)

# Create a custom colormap that mimics the spectrogram colors
cmap = ListedColormap(plt.cm.viridis(np.linspace(0.25, 1, 256))

# Plot the exponential curve using the custom colormap
plt.figure(figsize=(8, 4))
plt.scatter(t, exponential_curve, c=exponential_curve, cmap=cmap, s=10)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Exponential Curve')
plt.colorbar(orientation='vertical', label='Amplitude', aspect=50)

plt.show()
```
