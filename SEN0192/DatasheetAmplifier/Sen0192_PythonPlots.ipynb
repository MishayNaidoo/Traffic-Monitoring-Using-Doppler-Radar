## Minimal Motion Data


```python
# Read the data from the text file
with open('Minimal_Motion.txt', 'r') as file:
    data = file.read().strip().split(',')

# Convert the data to a NumPy array
data = np.array(data, dtype=int)
data = data/1024 * 5
data = data - 2.6
data = data[data>0]
# Create a time array (assuming each data point corresponds to a time step)


time = np.arange(len(data))

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(time, data, label='Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Data Plot')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Define the filter parameters
cutoff_frequency = 50  # Cutoff frequency in Hz
sampling_frequency = 15000 # Sampling frequency in Hz

# Normalize the cutoff frequency
normalized_cutoff = cutoff_frequency / (0.5 * sampling_frequency)

# Filter order (adjust as needed)
filter_order = 4

# Design the Butterworth low-pass filter
b, a = butter(filter_order, normalized_cutoff, btype='low')

# Generate a test signal (for demonstration purposes)
test_signal = data

# Apply the filter to the test signal
filtered_signal = lfilter(b, a, test_signal)

# Plot the original and filtered signals
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(test_signal)
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(filtered_signal)
plt.title('Filtered Signal')
plt.xlabel('Time (s)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Some Motion Data

```python
# Read the data from the text file
with open('Some_Motion.txt', 'r') as file:
    data = file.read().strip().split(',')

# Convert the data to a NumPy array
data = np.array(data, dtype=int)
data = data/1024 * 5
# Create a time array (assuming each data point corresponds to a time step)


time = np.arange(len(data))

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(time, data, label='Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Data Plot')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Number of data points
N = len(data)

# Calculate the sample spacing (assuming uniform sampling)
dt = 1/15000  # Replace with the actual time spacing between samples if known

# Perform FFT
data_fft = np.fft.fft(data)
freqs = np.fft.fftfreq(N, dt)  # Frequency values corresponding to the FFT components

# Plot the original data and its FFT
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(np.arange(N) * dt, data)
plt.title('Original Sampled Data')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(freqs, np.abs(data_fft))
plt.title('FFT of Sampled Data')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.xlim(0, 200)  # Display only positive frequencies
plt.ylim(0,10000)

plt.tight_layout()
plt.show()
```
